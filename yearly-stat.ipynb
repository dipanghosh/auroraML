```python
import pandas as pd

df = pd.read_csv("2018.csv")
```

```python
df['sum'] = df.iloc[:, 1:].sum(axis=1, skipna = True)
```

```python
df.sort_values('sum', ascending=False)
```

```python
df["count"] = df.count(axis='columns') - 4
df["fraction"] = df['sum']/df['count']
```

```python
df = df.sort_values('sum', ascending=False)
df.to_csv("2018_processed.csv")
```
